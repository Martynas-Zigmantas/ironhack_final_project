```python
airline_code_mapping = {
    'VL': 'MED-VIEW AIRLINE',
    'IV': 'WINDJET',
    'BQ': 'BLUEBIRD AIRWAYS',
    'VF': 'VALUJET AIRLINES',
    'OG': 'ONE-TWO-GO AIRLINES',
    'N0': 'NORSE ATLANTIC AIRWAYS',
    '6K': 'GERMANWINGS',
    'TLJ': 'JETLINE INTERNATIONAL',
    'OE': 'LAUDA EUROPE',
    'XY': 'FLYNAS',
    # Add more mappings here if needed
}

# Modified function to clean airline names, now accounting for airline codes and all caps
def clean_airline_name(airline):
    if pd.isna(airline):
        return np.nan
    # Remove airline codes in parentheses and convert name to uppercase
    clean_name = airline.split('(')[0].strip().upper()  
    # If the airline name is just a code (e.g., '(VL)'), map it to the actual airline name in all caps
    if clean_name == '':
        code = airline.strip('()')  # Extract the code
        return airline_code_mapping.get(code, np.nan)  # Return the mapped airline name or NaN if not found
    return clean_name

# Apply the updated cleaning function to the DataFrame
data['cleaned_airline'] = data['airline'].apply(clean_airline_name)
```

```python
def fill_airline_type(airline_type):
    if airline_type.lower() == 'unknown':  # Check if the value is 'unknown' (case-insensitive)
        return 'budget'
    return airline_type  # Return the original value if it's not 'unknown'

# Apply the function to the 'airline_type' column
data['airline_type'] = data['airline_type'].apply(fill_airline_type)
```

```python
def clean_airline_name(airline):
    if pd.isna(airline):
        return np.nan
    # Remove airline codes in parentheses and convert name to uppercase
    clean_name = airline.split('(')[0].strip().upper()  
    # If the airline name is just a code (e.g., '(VL)'), map it to the actual airline name
    if clean_name == '':
        code = airline.strip('()')  # Extract the code
        mapped_name = airline_code_mapping.get(code, np.nan)  # Get the mapped airline name or NaN
        if isinstance(mapped_name, str):  # Ensure it's a string before applying .title()
            return mapped_name.title()
        else:
            return np.nan  # Return NaN if not found in mapping
    return clean_name.title()  # Return title case for other cleaned names

# Apply the updated cleaning function to the DataFrame
data['cleaned_airline'] = data['airline'].apply(clean_airline_name)
```
